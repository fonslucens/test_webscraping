```python
imformations = [] # or list()
for content in contents:
    company = content.select('span > a[class="strong"]')[0]
    title = content.select('span > a[slot-sort*="19439151"]')[0]
    date = content.select('div[class="ddays"]')[0]
    print(company.text.strip(), title.text.strip(), date.text.strip())
    imformations.append([company.text.strip(), title.text.strip(), date.text.strip()])
imformations
```

```python
import pandas as pd
```

```python
pd_data = pd.DataFrame(imformations, columns=['company', 'title','date'])
pd_data
```

```python
pd_data.to_excel('./saves/companylist.xlsx')
```
