```python
browser.get('https://github.com/fonslucens/test_webscraping')
```

```python
from bs4 import BeautifulSoup
```

```python
html = BeautifulSoup(browser.page_source, 'html.parser')
html
```

```python
tags = html.select('div[role="row"].Box-row')
type(tags), len(tags)
```

```python
tag = tags[0]
tag
```

```python
name = tag.select('a[class="js-navigation-open Link--primary"]')[0]
name
```

```python
desc = tag.select('a[class="Link--secondary"]')[0]
desc
```

```python
time = tag.select('div > time-ago')[0]
time
```

```python
name.text.strip(), desc.text.strip(), time.text.strip()
```

```python
files = [] # or list()
for tag in tags:
    name = tag.select('a[class="js-navigation-open Link--primary"]')[0]
    desc = tag.select('a[class="Link--secondary"]')[0]
    time = tag.select('div > time-ago')[0]
    print(name.text.strip(), desc.text.strip(), time.text.strip())
    files.append([name.text.strip(), desc.text.strip(), time.text.strip()])
files
```

```python
import pandas as pd
```

```python
pd_data = pd.DataFrame(files, columns=['name', 'desc', 'time'])
pd_data
```

```python
pd_data.to_excel('./saves/github_scraping.xlsx')
```
